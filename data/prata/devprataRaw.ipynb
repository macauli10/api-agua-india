```python
colunas_para_converter = ['condutividade_min', 'bod_min_mg_l']

for coluna in colunas_para_converter:
    if coluna in df.columns:
        df[coluna] = pd.to_numeric(df[coluna], errors='coerce')

# Verificar os tipos de dados novamente
print("Tipos de dados após conversão:")
print(df.dtypes)
```

```python
def adicionar_medias_seguras(df):
    """
    Adiciona colunas de média apenas quando ambos valores (min e max) estão disponíveis
    """
    # Parâmetros físico-químicos (média aritmética)
    parametros_aritmeticos = {
        'temperatura': ('temperatura_min_c', 'temperatura_max_c', 'temperatura_media_c'),
        'oxigenio_dissolvido': ('oxigenio_dissolvido_min', 'oxigenio_dissolvido_max', 'oxigenio_dissolvido_medio'),
        'ph': ('ph_min', 'ph_max', 'ph_medio'),
        'condutividade': ('condutividade_min', 'condutividade_max', 'condutividade_media'),
        'bod': ('bod_min_mg_l', 'bod_max_mg_l', 'bod_medio_mg_l'),
        'nitrato': ('nitrato_min_mg_l', 'nitrato_max_mg_l', 'nitrato_medio_mg_l')
    }
    
    for param, (min_col, max_col, media_col) in parametros_aritmeticos.items():
        if min_col in df.columns and max_col in df.columns:
            mask = df[min_col].notna() & df[max_col].notna()
            df[media_col] = np.nan  # Inicializar com NaN
            df.loc[mask, media_col] = (df.loc[mask, min_col] + df.loc[mask, max_col]) / 2
    
    # Parâmetros microbiológicos (média geométrica - mais apropriada)
    parametros_geometricos = {
        'coliformes_fecais': ('coliformes_fecais_min', 'coliformes_fecais_max', 'coliformes_fecais_medio'),
        'coliformes_totais': ('coliformes_totais_min', 'coliformes_totais_max', 'coliformes_totais_medio'),
        'fecal': ('fecal_min', 'fecal_max', 'fecal_medio')
    }
    
    for param, (min_col, max_col, media_col) in parametros_geometricos.items():
        if min_col in df.columns and max_col in df.columns:
            mask = df[min_col].notna() & df[max_col].notna()
            df[media_col] = np.nan  # Inicializar com NaN
            # Calcular média geométrica apenas onde ambos valores são positivos
            mask_positivos = mask & (df[min_col] > 0) & (df[max_col] > 0)
            df.loc[mask_positivos, media_col] = np.sqrt(df.loc[mask_positivos, min_col] * df.loc[mask_positivos, max_col])
    
    return df

# Aplicar a função para calcular médias
df = adicionar_medias_seguras(df)

# Verificar resultados
print("\nPrimeiras linhas com médias calculadas:")
print(df[['local_monitoramento', 'ano', 'temperatura_media_c', 'ph_medio', 'coliformes_fecais_medio']].head())

# Estatísticas de completude
print("\nCompletude dos dados após cálculo de médias:")
colunas_media = [col for col in df.columns if 'medio' in col or 'media' in col]
for col in colunas_media:
    if col in df.columns:
        total = len(df)
        na_count = df[col].isna().sum()
        percent_na = (na_count / total) * 100
        print(f"{col}: {total - na_count}/{total} ({100 - percent_na:.1f}%) preenchidos")

# Análise por tipo de corpo d'água
if 'tipo_corpo_agua' in df.columns:
    print("\nCompletude por tipo de corpo d'água:")
    for tipo in df['tipo_corpo_agua'].unique():
        subset = df[df['tipo_corpo_agua'] == tipo]
        print(f"\n{tipo}:")
        for col in ['temperatura_media_c', 'ph_medio', 'coliformes_fecais_medio']:
            if col in df.columns:
                completude = subset[col].notna().mean() * 100
                print(f"  {col}: {completude:.1f}%")

# Salvar o dataset com as médias
df.to_csv('aguaPrataMediasRaw.csv', index=False)
print("\nDataset salvo como 'aguaPrataMediasRaw.csv'")
```
