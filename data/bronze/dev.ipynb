```python
colunas_traduzidas = {
    "STN code": "codigo_stn",
    "Monitoring Location": "local_monitoramento",
    "Year": "ano",
    "Type Water Body": "tipo_corpo_agua",
    "State Name": "estado",
    "Temperature (C) - Min": "temperatura_min_c",
    "Temperature (C) - Max": "temperatura_max_c",
    "Dissolved - Min": "oxigenio_dissolvido_min",
    "Dissolved - Max": "oxigenio_dissolvido_max",
    "pH - Min": "ph_min",
    "pH - Max": "ph_max",
    "Conductivity (Â¬Âµmho/cm) - Min": "condutividade_min",
    "Conductivity (Â¬Âµmho/cm) - Max": "condutividade_max",
    "BOD (mg/L) - Min": "bod_min_mg_l",
    "BOD (mg/L) - Max": "bod_max_mg_l",
    "NitrateN (mg/L) - Min": "nitrato_min_mg_l",
    "NitrateN (mg/L) - Max": "nitrato_max_mg_l",
    "Fecal Coliform (MPN/100ml) - Min": "coliformes_fecais_min",
    "Fecal Coliform (MPN/100ml) - Max": "coliformes_fecais_max",
    "Total Coliform (MPN/100ml) - Min": "coliformes_totais_min",
    "Total Coliform (MPN/100ml) - Max": "coliformes_totais_max",
    "Fecal - Min": "fecal_min",
    "Fecal - Max": "fecal_max"
}
```

```python
df = df.rename(columns=colunas_traduzidas)
```

```python

cols_para_converter = [
    "oxigenio_dissolvido_min", "oxigenio_dissolvido_max",
    "nitrato_min_mg_l", "nitrato_max_mg_l",
    "coliformes_fecais_min", "coliformes_fecais_max",
    "coliformes_totais_min", "fecal_min", "fecal_max"
]

for col in cols_para_converter:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# colunas para substituir NaN por 0
colunas_para_zero = [
    "temperatura_min_c", "temperatura_max_c",
    "oxigenio_dissolvido_min", "oxigenio_dissolvido_max",
    "ph_min", "ph_max",
    "condutividade_min", "condutividade_max",
    "bod_min_mg_l", "bod_max_mg_l",
    "nitrato_min_mg_l", "nitrato_max_mg_l",
    "coliformes_fecais_min", "coliformes_fecais_max",
    "coliformes_totais_min", "coliformes_totais_max",
    "fecal_min", "fecal_max"
]

# cria uma cÃ³pia tratada
df_tratado = df.copy()

# garantir tipos numÃ©ricos
df_tratado[colunas_para_zero] = df_tratado[colunas_para_zero].apply(pd.to_numeric, errors="coerce")

# preencher NaN com 0
df_tratado[colunas_para_zero] = df_tratado[colunas_para_zero].fillna(0)

# salva no prata como versÃ£o tratada
df_tratado.to_csv("../prata/agua_prata_tratado.csv", index=False, encoding="utf-8")
```
